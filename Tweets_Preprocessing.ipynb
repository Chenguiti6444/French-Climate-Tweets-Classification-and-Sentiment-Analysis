```python
!pip install pyspellchecker
!pip install spacy
!python -m spacy download fr_core_news_sm
import pandas as pd
import re
from spellchecker import SpellChecker
import nltk
nltk.download('punkt')
nltk.download('stopwords')
from nltk.corpus import stopwords
import spacy

# Charger le modèle français de spacy
nlp = spacy.load('fr_core_news_sm')
from google.colab import drive

# Mount Google Drive
drive.mount('/content/drive', force_remount=True)
```

```python
# Load data from the CSV files
df_train = pd.read_csv('/content/drive/MyDrive/Projet Intégré S8 (Tweets Réchauffement Climatique)/train.csv')
df_test = pd.read_csv('/content/drive/MyDrive/Projet Intégré S8 (Tweets Réchauffement Climatique)/test.csv')

# Initialize the spell checker for French
spell = SpellChecker(language='fr')

def preprocess_text(text):

    # Conversion en minuscules
    text = text.lower()

    # Suppression des URL
    text = re.sub(r"http\S+", "",text)

    # Suppression des mentions utilisateur (usernames)
    text = re.sub(r"@\w+", "", text)

    # Suppression de la ponctuation et des caractères spéciaux sauf les caractères spéciaux français
    text = re.sub(r"[^\w\séèêàôîûç]+|[\d]+|[_]+", " ", text, flags=re.UNICODE)

    # Tokenisation avec spaCy
    doc = nlp(text)
    words = [token.text for token in doc]

    """# Correction orthographique
    corrected_words = []
    for word in words:
        corrected_word = spell.correction(word)
        if corrected_word is not None:
            corrected_words.append(corrected_word)
        else:
            corrected_words.append(word)  # Ajouter le mot original s'il n'y a pas de correction"""

    # Suppression des stop words
    stop_words = set(stopwords.words("french"))
    words = [word for word in words if word not in stop_words]

    # Suppression des lettres solitaires
    words = [word for word in words if len(word) > 1]

    # Lemmatisation avec spaCy
    doc = nlp(" ".join(words))
    words = [token.lemma_ for token in doc]

    words = [word if word != "réchauffemer" else "réchauffement" for word in words]

    # Rejoindre les mots en texte
    text = " ".join(words)
    return text

# Preprocess each tweet in the training dataset
df_train['Preprocessed_Tweet'] = df_train['Tweet'].apply(preprocess_text)

# Preprocess each tweet in the testing dataset
df_test['Preprocessed_Tweet'] = df_test['Tweet'].apply(preprocess_text)
```
